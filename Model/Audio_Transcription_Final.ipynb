**Import the Necessary Libraries**
**Currently only 3 Libraries are required**

1.   Librosa to load the audio file
2.   transformers to use the required model
3.   IPython.display to listen to the audio file (not necessary can be removed if not required)



```
# Model file .bin or .tf or .pt, model weights, requirements.txt, model code, inference code for smooth deployment to GCP with Docker
import librosa
from IPython.display import Audio
from transformers import AutoProcessor,WhisperForConditionalGeneration
```

**Import the Necessary Model**


1.   We are going to use WhisperForConditionalGeneration as the based model
2.   We also require the AutoProcessor for the base model



```
                                                                        #large , medium , large-v1,v2,v3
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-base.en")
processor = AutoProcessor.from_pretrained("openai/whisper-base.en")
```

**Read the audio file using librosa**

1.   First argument has the returned array from the librosa.load function
2.   Second argument has the returned sample_rate . We are keeping the sample rate at 16000



```
audio,sr=librosa.load("/content/zaraar.m4a",sr=16000)
```

**Following is the inferrence code**


1.   "inputs" will prepare the audio by passing it through the processor
2.   "generated_ids" will get the generated ids by passing inputs to the model
3.   "transcription" will decode the generated ids by passing them back to the processor


```
inputs = processor(audio, return_tensors="pt",sampling_rate=sr)
generated_ids = model.generate(**inputs)
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)
```

**Given Below is the inferrence code**


1.   *First function is to load the pretrained model*

        *   First Arugment is the file path of the saved processor
        *   Second Argument is the file path of the saved model
        *   It will return the model and the processor

2.   *Second Function is to make inferrence*

        *    First argument is the return model from First Function
        *    Second Argument is the returned processor from the First Function
        *    Third argument is the test audio file path
        *    Transcription of the audio file will be returned









```
def load_pretrained(processor_file_path,model_file_path):
  model_1=WhisperForConditionalGeneration.from_pretrained(model_file_path)
  processor_1=AutoProcessor.from_pretrained(processor_file_path)

  return model_1,processor_1

def make_inferrence(Model,Processor,audio_path):
  import warnings
  warnings.filterwarnings("ignore")

  audio,sr=librosa.load(audio_path,sr=16000)
  inputs = Processor(audio, return_tensors="pt",sampling_rate=sr)
  generated_ids = Model.generate(**inputs)
  transcription = Processor.batch_decode(generated_ids, skip_special_tokens=True)

  return transcription[0]
```

```
model,processor=load_pretrained("/content/model_processor","/content/main_model")
```

```
make_inferrence(model,processor,"/content/zaraar.m4a")
```
